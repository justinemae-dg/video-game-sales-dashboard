```python
import pandas as pd

df = pd.read_csv("vgsales.csv")

df
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.dropna(subset=["Publisher"]) # removes rows with missing Publisher
```

```python
# Remove non-numeric entries and missing values
df["Year"] = pd.to_numeric(df["Year"], errors="coerce") # "coerce" converts invalid text to NaN
```

```python
df = df.dropna(subset=["Year"]) # drops rows where year is NaN
```

```python
df["Year"] = df["Year"].astype(int) # converts to integer safely
```

```python
# standardize text columns
df["Name"] = df["Name"].str.strip()
df["Platform"] = df["Platform"].str.upper()
df["Genre"] = df["Genre"].str.title()
df["Publisher"] = df["Publisher"].str.title()
```

```python
# aggregated datasets
# Aggregation - combining multiple rows into a single row (e.g., all rows having year 2005 will be combined)
sales_by_year = df.groupby("Year")["Global_Sales"].sum().reset_index() # shows total global sales trend over time
sales_by_year
```

```python
top_games = df.sort_values(by="Global_Sales", ascending=False).head(10) # sort in greatest to smallest
top_games
```

```python
sales_by_genre = df.groupby("Genre")["Global_Sales"].sum().reset_index()
sales_by_genre
```

```python
# save to csv
df.to_csv("vgsales_clean.csv", index=False)
sales_by_year.to_csv("sales_by_year.csv", index=False)
top_games.to_csv("top_games.csv", index=False)
sales_by_genre.to_csv("sales_by_genre.csv", index=False)
```
